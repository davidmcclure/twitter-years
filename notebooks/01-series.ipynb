```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
import numpy as np

from twitter_years.models import Year
```

```python
counts = Year.year_counts()
```

```python
plt.figure(figsize=(20, 10))

plt.xlim((1900, 2000))
plt.xticks(np.arange(1900, 2000, 5))

plt.hist(
    list(counts.keys()),
    bins=len(counts),
    weights=list(counts.values()),
    edgecolor='black',
    zorder=3,
)

plt.show()
```
